```python
import os
import pandas as pd
from jinja2 import Environment, FileSystemLoader

# Load the CSV file into a DataFrame
books_df = pd.read_csv(r'C:\Users\dexter\Downloads\updated1.csv')

# Set up the Jinja2 environment
env = Environment(loader=FileSystemLoader(r'C:\Users\dexter\Downloads\generate'))

# Fetch the new template file
template = env.get_template('template3.html')

# Define the output directory
output_dir = r'C:\Users\dexter\Downloads\output_directory'

# Check if the output directory exists, create it if it doesn't
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Group data by address
grouped = books_df.groupby('Address of the Book Box')

# Generate a subpage for each group
for address, group in grouped:
    # Since latitude and longitude are the same for the same address, we take the first entry
    latitude = group['Latitude'].iloc[0]
    longitude = group['Longitude'].iloc[0]

    # Collect book details into a list of dictionaries
    books = [{
        'title': row['Title of the Book'],
        'author': row['Name of the First Author or Publisher'],
        'book_url': row['Book URL'],
        'image_link': row['Image Link']
    } for index, row in group.iterrows()]

    # Render the template with the grouped data
    rendered_html = template.render(
        address=address,
        latitude=latitude,
        longitude=longitude,
        books=books
    )
    
    # Create a valid filename from the address, replacing '/' with '_' to avoid path issues
    filename = f"{address.replace('/', '_')}.html"
    
    # Construct the full file path
    file_path = os.path.join(output_dir, filename)

    # Write the output to a file
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(rendered_html)
```
